```python
def remove_whitespaces(input):
    """
    Returns text without extra whitespaces
    """
    return " ".join(input.split())
  
def remove_html_tags(input):
    """
    Returns text without HTML tags
    """
    soup = BeautifulSoup(input, "html.parser")
    stripped_input = soup.get_text(separator=" ")
    return stripped_input    
```

```python
import googletrans  # ensure version 3.1 to enable bulk translate
from googletrans import Translator
translator = Translator()

# print(googletrans.LANGUAGES)  # display all translatable languages
```

```python
# batch comments and feed to googletrans (15K char limit per time)

batch_size = 150
batch_half = int(batch_size/2)

commentEn = list()
counter=0

for i in range(0, len(nonEn_idx), batch_size):
    transText = [str(reviewHk.comments[i]) for i in nonEn_idx[i:i+batch_size]]
    if sum(len(char) for char in transText) > 14500:
        translations = translator.translate([str(reviewHk.comments[i]) for i in nonEn_idx[i:i+batch_half]], dest='en')
        for trans in translations: 
            commentEn.append(trans.text.lower().replace('\r<br/>',' ').replace('<br/>',' '))
        translations = translator.translate([str(reviewHk.comments[i]) for i in nonEn_idx[i+batch_half:i+batch_size]], dest='en')
        for trans in translations: 
            commentEn.append(trans.text.lower().replace('\r<br/>',' ').replace('<br/>',' '))
    else: 
        translations = translator.translate(transText, dest='en')
        for trans in translations: 
            commentEn.append(trans.text.lower().replace('\r<br/>',' ').replace('<br/>',' '))
    counter+=1
    print(counter)
len(commentEn)
```

### Translate Listing Descriptions

```python
listingHk = pd.read_csv('./data/listingHk.csv')
```

```python
# convert to lower and remove headings,ie 'the space', 'guest access', 'other things to note'
#listingHk.description = listingHk.description.apply(lambda x : str(x).lower())
listingHk.description = listingHk.description.apply(lambda x : str(x).replace('<b>the space</b>',''))
listingHk.description = listingHk.description.apply(lambda x : str(x).replace('<b>guest access</b>',''))
listingHk.description = listingHk.description.apply(lambda x : str(x).replace('<b>other things to note</b>',''))
```

```python
listingHk.description = listingHk.description.apply(lambda x : remove_whitespaces(remove_html_tags(x)))
```

```python
listingHk.description[63]
```

```python
#  Get all non-english rows

nonEn_idx = list()

for i in range(len(listingHk.description)):
    if nlp(str(listingHk.description[i]))._.language.get('language') != 'en':
        nonEn_idx.append(listingHk.index[i])
        print(i)
```

```python
len(nonEn_idx)
```

```python
# batch comments and feed to googletrans (15K char limit per time)

batch_size = 150
batch_half = int(batch_size/2)

descEn = list()
counter=0

for i in range(0, len(nonEn_idx), batch_size):
    transText = [str(listingHk.description[i]) for i in nonEn_idx[i:i+batch_size]]
    if sum(len(char) for char in transText) > 14500:
        translations = translator.translate([str(listingHk.description[i]) for i in nonEn_idx[i:i+batch_half]], dest='en')
        for trans in translations: 
            descEn.append(trans.text)
        translations = translator.translate([str(listingHk.description[i]) for i in nonEn_idx[i+batch_half:i+batch_size]], dest='en')
        for trans in translations: 
            descEn.append(trans.text)
    else: 
        translations = translator.translate(transText, dest='en')
        for trans in translations: 
            descEn.append(trans.text)
    counter+=1
    print(counter)
len(descEn)
```

```python
df = pd.DataFrame(descEn, index=nonEn_idx, columns=['descEn'])
#df.to_csv('./data/descEn.csv')
```

```python
listingHk = pd.merge(listingHk, df, how='outer', left_index=True, right_index=True)
```
